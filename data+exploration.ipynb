```python
df = pd.read_csv('./data/labels.csv')
df.head()
```

```python
cls = set(df['breed'])
print('all dogs classes: \n{}'.format(cls))
```

```python
print('total class number:')
print(len(cls))
```

```python
all_dog = df.values
dog_num = dict((i, 0) for i in cls)
for item in all_dog:
    dog_num[item[1]] += 1
```

### split data into train and validaiton

```python
idx2cls = dict((i, j) for i, j in enumerate(cls))
cls2idx = dict((j, i) for i, j in enumerate(cls))
```

```python
idx2cls
```

```python
with open('./data/idxdic.json', 'w') as f:
    json.dump(idx2cls, f)
```

```python
cls2idx
```

```python
with open('./data/clsdic.json', 'w') as f:
    json.dump(cls2idx, f)
```

```python
with open('./data/labels.csv', 'r') as f:
    img_list = [i.strip().split(',') for i in f.readlines()[1:]]
```

```python
shuffle(img_list)
valid_num = dict((i, 0) for i in range(120))
train_list = []
valid_list = []
for im in img_list:
    im[1] = cls2idx[im[1]]
    if valid_num[im[1]] < 7:
        valid_list.append(im)
        valid_num[im[1]] += 1
    else:
        train_list.append(im)
```

```python
print('train num: {}, valid num: {}, train_valid num: {}'.format(len(train_list), len(valid_list), len(img_list)))
```

```python
with open('./data/train.json', 'w') as f:
    json.dump(train_list, f)
with open('./data/valid.json', 'w') as f:
    json.dump(valid_list, f)
with open('./data/train_valid.json', 'w') as f:
    json.dump(img_list, f)
```
